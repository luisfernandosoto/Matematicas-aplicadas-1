```python
import matplotlib.pyplot as plt
import numpy as np
velocidad_inicial= 30.0
gravedad= 9.8
def trayectoria(angulo):
    angulo_rad= np.radians(angulo)
    value= (2 * velocidad_inicial * np.sin(angulo_rad)) / gravedad
    value_2= np.linspace(0, value, 100)

    x= velocidad_inicial * np.cos(angulo_rad) * value_2
    y= velocidad_inicial * np.sin(angulo_rad) * value_2 - 0.5 * gravedad * value_2**2
    return x, y

angulos= [30, 45, 60, 55]

for angulo in angulos:
    x, y= trayectoria(angulo)
    plt.plot(x, y, color="green")

plt.title("Movimiento parabolico")
plt.xlabel("distancia recorrida del proyectil")
plt.ylabel("altura que alcanza el proyectil")
plt.grid(color="gray")
plt.show()
```
