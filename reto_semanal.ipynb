```python
import matplotlib.pyplot as plt
import numpy as np
velocidad_inicial= 30.0
gravedad= 9.8
def trayectoria(angulo):
    angulo_rad= np.radians(angulo)
    value= (2 * velocidad_inicial * np.sin(angulo_rad)) / gravedad
    value_2= np.linspace(0, value, 100)

    x= velocidad_inicial * np.cos(angulo_rad) * value_2
    y= velocidad_inicial * np.sin(angulo_rad) * value_2 - 0.5 * gravedad * value_2**2
    return x, y

angulos= [30, 45, 60, 55]

for angulo in angulos:
    x, y= trayectoria(angulo)
    plt.plot(x, y, color="green")

plt.title("Movimiento parabolico")
plt.xlabel("distancia recorrida del proyectil")
plt.ylabel("altura que alcanza el proyectil")
plt.grid(color="gray")
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

def generar_puntos_sierpinski(num_puntos):
    """Genera puntos del tringulo de Sierpinski usando el mtodo aleatorio."""

    vertices = [(0, 0), (1, 0), (0.5, (3**0.5) / 2)]


    x, y = 0.5, (3**0.5) / 6
    puntos = []

    for _ in range(num_puntos):

        vertice = random.choice(vertices)

        x = (x + vertice[0]) / 2
        y = (y + vertice[1]) / 2
        puntos.append((x, y))

    return puntos

def dibujar_fractal_sierpinski(num_puntos):
    """Dibuja el fractal de Sierpinski usando matplotlib."""
    puntos = generar_puntos_sierpinski(num_puntos)

    plt.title(f'Fractal de Sierpinski con {num_puntos} puntos')
    plt.xlabel("eje x")
    plt.ylabel("eje y")
    plt.scatter(*zip(*puntos), s=0.1, c='red')
    plt.grid(color="gray")

    plt.show()


dibujar_fractal_sierpinski(10000)
```
